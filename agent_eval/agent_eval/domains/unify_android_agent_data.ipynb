```python
import json
import os
```

```python
# src_path = "/home/jiayipan/data/ios_traj/output_cogagent"
# src_path = "/home/jiayipan/data/GUI_Proj/output_autoui_large"
# src_path = "/home/jiayipan/data/android_offline/output_autoui_base_300"
src_path = "/home/jiayipan/data/android_offline/output_autoui_base_filteredbc_test"
full_data = []
success_log = json.load(open(os.path.join(src_path, "log.json"), "r"))
success_trajs = []
for task_name, task_log in success_log.items():
    for traj_name, traj_log in task_log.items():
        if traj_log == "SUCCESS":
            success_trajs.append(traj_name + ".json")
for file in os.listdir(src_path):
    if file in success_trajs:
        print(file)
        with open(os.path.join(src_path, file), "r") as f:
            data = json.load(f)
        full_data.append(data)
```

```python
all_trajectory_log = []
for traj in full_data:
    uid  = traj[0]['obs'].split("/")[-1][:-6]
    goal = traj[0]['task']
    traj_log = {
        "uid": uid,
        "intent": goal,
        "response": None,
        "steps": []
    }
    for step in traj:
        step_log = {
            "img": step['obs'].split("/")[-1],
            "other": step
        } 
        traj_log['steps'].append(step_log)
    all_trajectory_log.append(traj_log)
```
